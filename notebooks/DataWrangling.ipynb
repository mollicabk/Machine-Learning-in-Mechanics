# Importando bibliotecas


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Sistema

```python
df_S = pd.read_excel('sistema.xlsx', index_col=None, header=None)
df_S = df_S.T

df_S = pd.DataFrame(df_S)
df_S.columns = df_S.iloc[0]
df_S = df_S[1:]
df_S = df_S.reset_index(drop=True)

df_S
```

```python
df_S.info()
```

```python
sns.heatmap(df_S.isnull())
```

```python
for col in df_S.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df_S[col], kde=True, bins = 15)
    plt.title(f'Distribuio de {col}')
    plt.show()
```

```python
df_aco =(df_S[df_S['massa']<400]['massa']).astype(float).plot(kind='hist', bins=15,edgecolor='white')
```

# Tempos

```python
df_T = pd.read_excel('tempos.xlsx')
df_T
```

```python
df_T.info()
```

```python
df_1 = df_T[['tc_aluminio', 'tc_aco']]
sns.heatmap(df_1.isnull())
```

```python
df_1_melted = df_1.melt(var_name='tempo', value_name='Tempo')
df_1_melted
```

```python
df_T.head()
```

```python
melted_df = pd.melt(df_T, id_vars=['groups'], value_vars=['tc_aluminio', 'tc_aco'], var_name='tempo', value_name='Tempo')

sns.histplot(data=melted_df, x='Tempo', hue='groups', kde=True)
plt.title('Distribuio de Tempo por Grupo')
plt.show()
```

```python
sns.boxplot(x='groups', y='Tempo', data=melted_df, hue = 'groups')
plt.title('Boxplot of Time by Group')
plt.xlabel('Group')
plt.ylabel('Time')
plt.show()
```

```python
sns.boxplot(x='groups', y='tc_aluminio', data=df_T, hue = 'groups')
plt.title('ALUMINIO')
plt.xlabel('Group')
plt.ylabel('Time')
plt.show()


sns.boxplot(x='groups', y='tc_aco', data=df_T, hue = 'groups')
plt.title('AO')
plt.xlabel('Group')
plt.ylabel('Time')
plt.show()
```
